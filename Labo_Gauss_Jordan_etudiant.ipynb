Autrement dit, apr√®s cette op√©ration, l‚Äôentr√©e $A_{i,\text{ col}}$ doit devenir $0$.

On utilise la formule :


$L_j <- L_j - p \cdot L_j \quad$ o√π $p$ est l'√©l√©ment en position $\text{col}$ de la ligne $i$.

Hypoth√®ses importantes :

- La ligne `pivot_row` contient d√©j√† un pivot √©gal √† **1** dans la colonne `col`.
- On ne modifie pas la ligne pivot.
- On modifie directement la matrice `A`.

#### Exemple : pivot sur la 2·µâ ligne et √©limination **au-dessus** de la ligne pivot

$A =

\end{bmatrix}$

- le pivot est en colonne 2 (indice `col = 1`) et vaut d√©j√† **1**,

‚úÖ √âtape 4 ‚Äî Trouver une ligne pivot dans une colonne

üîç **But :** √Ä partir d'une certaine ligne de la matrice, on veut trouver la ligne avec une entr√©e non-nulle dans une colonne donn√©e.

‚úÖ Je me suis occuper de faire de cette √©tape dans les lignes de codes suivantes.


```python
def trouver_pivot(M,ligne_courante,col):
    # On trouve le num√©ro de ligne du pivot dans la "col"-i√®me colonne
    ligne_pivot = None
    for r in range(ligne_courante, M.shape[0]):
        if abs(M[r, col]) > 1e-12:
            ligne_pivot = r
            return ligne_pivot
    return ligne_pivot
```

Voici les √©tapes de l'algorithme pour chaque colonne.
1. Chercher un pivot dans la colonne.
2. S‚Äôil existe :
   - mettre la ligne pivot au bon endroit (√©change),
   - diviser la ligne pour que le pivot devienne 1,
   - √©liminer la colonne dans toutes les autres lignes. **Ceci est la seule ligne qui reste √† modifier.**
3. Passer √† la colonne suivante.

L‚Äôobjectif est de transformer progressivement la matrice augment√©e :

$[A \mid b]$

en une matrice r√©duite :


$[I \mid x]$

o√π $ x $ est la solution du syst√®me.

Vous compl√®terez la partie manquante dans la cellule de code ci-dessous, gr√¢ce aux fonctions que vous avez √©crites pr√©c√©demment.


```python
def gauss_jordan_reels(A, b, show_steps=False):
    """R√©soudre Ax = b en utilisant l'algorithme de Gauss‚ÄìJordan.
    Retourne une solution x en tant qu'array Numpy 1D.
    Cr√©er une erreur s'il n'y a pas de solution.
    """
    A = np.array(A, dtype=float).copy()
    b = np.array(b, dtype=float).reshape(-1)
    n = A.shape[0]
    M = np.hstack([A, b.reshape(-1, 1)])  # matrice augment√©e par concat√©nation horizontale.

    if show_steps:
            print(f"Matrice initiale :")
            print_augmented_matrix_latex(M,b)

    row = 0
    for col in range(n):
        # On trouve le num√©ro de ligne du pivot dans la "col"-i√®me colonne
        ################################################################
        # Appel √† ma fonction de l'√©tape 4
        ligne_pivot = trouver_pivot(M, row, col)
        ################################################################

        # √©changer deux lignes si le pivot n'est pas sur la bonne ligne
        if ligne_pivot != row:
            ################################################################
            # Appel √† votre fonction de l'√©tape 2
            echanger_lignes(M, row, ligne_pivot)
            ligne_pivot = row
            ################################################################

        # Le pivot est en ligne "row", il est dans la colonne "col"
        # Diviser la ligne row par son pivot
        ################################################################
        # Appel √† votre fonction de l'√©tape 1
        diviser_ligne(M, ligne_pivot, M[ligne_pivot, col])
        ################################################################

        # √âlimination sur les autres lignes.
        for ligne_quelconque in range(n):
            # Pour toutes les lignes "r" sauf celle du pivot qui est "row"
            if ligne_quelconque != ligne_pivot:
                # Faites une √©limination ligne
                ################################################################
                # Appel √† votre fonction de l'√©tape 3
                # VOTRE CODE
                # id√©e: utilisez un appel √† votre fonction de l'√©tape 3 avec les
                # param√®tres appropri√©s
                pass
                ################################################################


        b = M[:, n]

        if show_steps:
            if col == 0: print(f"Apr√®s avoir trait√© le pivot de la 1√®re colonne:")
            else: print(f"Apr√®s avoir trait√© le pivot de la {col+1}e colonne:")
            print_augmented_matrix_latex(M[:,:-1],b)

        row += 1
        if row == n:
            break

    # Un petit message d'erreur si le syst√®me n'a pas de solution.
    A_rref = M[:, :n]
    b_rref = M[:, n]
    rank = np.sum(np.any(np.abs(A_rref) > 1e-9, axis=1))
    if rank < n:
        raise ValueError("Le syst√®me n'a pas de solution unique")

    return b_rref

#######################################################################
# Fonction pour imprimer des plus beaux trucs, ne pas toucher.

# Une fonction (g√©n√©r√©e avec l‚Äôaide d‚Äôune IA) qui permet d‚Äôafficher √©l√©gamment
# une matrice ou une matrice augment√©e sous forme LaTeX dans un notebook Jupyter.
def print_augmented_matrix_latex(A, b=None):
    """
    Affiche la matrice A ou la matrice augment√©e [A | b] en LaTeX.

    Param√®tres
    ----------
    A : array numpy de dimension (m √ó n)
        La matrice des coefficients.

    b : array numpy de longueur m, ou None
        Si b est None, seule la matrice A est affich√©e.
        Si b est fourni, la matrice augment√©e [A | b] est affich√©e.
    """

    A = np.array(A)
    m, n = A.shape

    # Cas 1 : Pas de colonne augment√©e, on affiche seulement A
    if b is None:
        rows = [
            " & ".join(str(A[i, j]) for j in range(n))
            for i in range(m)
        ]
        matrix_body = " \\\\ ".join(rows)
        latex = "\\left[\\begin{array}{%s}\n%s\n\\end{array}\\right]" % (
            "c" * n,
            matrix_body
        )
        return display(Math(latex))

    # Cas 2 : Matrice augment√©e [A | b]
    b = np.array(b).reshape(-1)
    if len(b) != m:
        raise ValueError("La longueur de b doit correspondre au nombre de lignes de A.")

    rows = []
    for i in range(m):
        left = " & ".join(str(A[i, j]) for j in range(n))
        right = str(b[i])
        rows.append(left + " & " + right)

    matrix_body = " \\\\ ".join(rows)
    latex = (
        "\\left[\\begin{array}{%s|c}\n%s\n\\end{array}\\right]"
        % ("c" * n, matrix_body)
    )
    display(Math(latex))
    print()
```
